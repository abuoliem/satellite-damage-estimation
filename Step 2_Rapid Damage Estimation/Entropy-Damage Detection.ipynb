```python
#################################################### SSR correction--pre-disaster  ###################################################
import cv2
import numpy as np
import os

def single_scale_retinex(image, sigma):
    """
    Apply Single-Scale Retinex (SSR) to an image.

    Parameters:
        image (numpy array): Input image (grayscale).
        sigma (float): Standard deviation of the Gaussian filter.

    Returns:
        numpy array: Retinex-corrected image.
    """
    # Estimate illumination using Gaussian blur
    illumination = cv2.GaussianBlur(image, (0, 0), sigma)

    # Prevent division by zero
    illumination[illumination == 0] = 1e-6

    # Compute reflectance (Retinex output)
    reflectance = np.log1p(image) - np.log1p(illumination)
    return reflectance

# Define input image path
image_path = r"C:\Users\abuoliem\Box\3.Third year2025\1.Fall25\3.Research\5.Phase1_Damage paper\D.google earth pro\Pre image_1-8-20\Location1\50x50\Lake Charles_Pre_1-8-20.jpg"

# Define output directory
output_directory = r"C:\Users\abuoliem\Box\3.Third year2025\1.Fall25\3.Research\5.Phase1_Damage paper\D.google earth pro\Pre image_1-8-20\Location1\50x50"

# Ensure output directory exists
os.makedirs(output_directory, exist_ok=True)

# Load a grayscale image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE).astype(np.float32) + 1e-6  # Avoid log(0)

# Apply Retinex for sigma = 30
sigma = 30
retinex_image = single_scale_retinex(image, sigma)

# Normalize reflectance for visualization
retinex_image_normalized = cv2.normalize(retinex_image, None, 0, 1, cv2.NORM_MINMAX)

# Define output file path
output_filename = os.path.join(output_directory, f"retinex_corrected_sigma_{sigma}.jpg")

# Save the Retinex-corrected image
cv2.imwrite(output_filename, (retinex_image_normalized * 255).astype(np.uint8))

print(f"Retinex-corrected image saved for sigma={sigma} as {output_filename}")
```

```python
#Grayscale Image Entropy (Normalized and Non-Normalized) of pre-disaster image (1-8-20)
import matplotlib.pyplot as plt #For visualizing the image and entropy values.
from PIL import Image #PIL (Python Imaging Library): For loading and processing the image.
import numpy as np #For numerical operations.
import os
def calculate_entropy(cell, normalize=False):
    """
    Calculate entropy for a given cell (image region).
    
    Parameters:
        cell (numpy array): The image region as a NumPy array.
        normalize (bool): Whether to normalize pixel values to the range [0, 1].
    
    Returns:
        float: Entropy of the cell.
    """
########################################### Step 2: Pixel Normalization  ###################################################
#By normalizing, you remove the impact of absolute brightness and ensure consistency across all calculations.
    # if normalize:
    cell = cell / 255.0  # Scale pixel intensity values to [0, 1]
    histogram, _ = np.histogram(cell.flatten(), bins=256, range=(0, 1))
    # else:
    #     histogram, _ = np.histogram(cell.flatten(), bins=256, range=(0, 256)) leaving the intenisty without normalization
    
    # Normalize the histogram to sum to 1//The histogram is normalized so the probabilities (𝑝𝑖) sum to 1//
    histogram = histogram / np.sum(histogram)
    histogram = histogram[histogram > 0]  # Exclude zero probabilities
    
    # Compute entropy
    entropy = -np.sum(histogram * np.log2(histogram))

########################################### Step 4: Entropy Normalization  ###################################################
    # Scale to [0, 1] for normalized case
    if normalize:
        entropy = entropy / np.log2(256)
    
    return entropy

########################################### Step 1: convert image to grayscale  ###################################################

def analyze_entropy(image_path, output_directory, cell_size_x, cell_size_y, resolution_x, resolution_y, visualize=False):
    """
    Analyze entropy for each cell in an image and optionally visualize the results.
    
    Parameters:
        image_path (str): Path to the image file.
        cell_size_x (float): Cell size in meters along the width.
        cell_size_y (float): Cell size in meters along the height.
        resolution_x (float): Spatial resolution in meters per pixel (X-axis).
        resolution_y (float): Spatial resolution in meters per pixel (Y-axis).
        visualize (bool): Whether to visualize the entropy grids and save the plots.
    
    Returns:
        tuple: Normalized and non-normalized entropy grids as NumPy arrays.
    """
    # Ensure output directory exists
    os.makedirs(output_directory, exist_ok=True)
    
    # Load the image and convert to grayscale
    image = Image.open(image_path).convert('L')  # Convert to grayscale
    image_array = np.array(image)
    
    # Calculate cell size in pixels
    cell_size_px_x = int(cell_size_x / resolution_x) #for 50x50m and resolution 0.07m/pixel then it is 714 pixels
    cell_size_px_y = int(cell_size_y / resolution_y)
    
    # Get image dimensions
    height, width = image_array.shape
    
    # Initialize Entropy Grids
    normalized_entropy_grid = []
    non_normalized_entropy_grid = []

########################################### Step 3: calculate entropy ############################################################
    # Divide the image into cells and calculate entropy
    for y in range(0, height, cell_size_px_y):
        norm_row = []
        non_norm_row = []
        for x in range(0, width, cell_size_px_x):
            # Extract the cell
            cell = image_array[y:y + cell_size_px_y, x:x + cell_size_px_x]
            
            # Calculate entropy
            entropy_non_normalized = calculate_entropy(cell, normalize=False)
            entropy_normalized = calculate_entropy(cell, normalize=True)
            
            # Append to rows
            norm_row.append(entropy_normalized)
            non_norm_row.append(entropy_non_normalized)
        
        normalized_entropy_grid.append(norm_row)
        non_normalized_entropy_grid.append(non_norm_row)
    
    # Convert entropy grids to NumPy arrays
    normalized_entropy_grid_pre = np.array(normalized_entropy_grid)
    non_normalized_entropy_grid_pre = np.array(non_normalized_entropy_grid)

########################################### Step 5: Visualization      ############################################################
    if visualize:
        # Visualize Original Image with Grid
        fig, ax = plt.subplots(figsize=(12, 12))
        ax.imshow(image_array, cmap='gray', extent=[0, width, height, 0])
        for y in range(0, height, cell_size_px_y):
            ax.axhline(y, color='cyan', linewidth=0.5)
        for x in range(0, width, cell_size_px_x):
            ax.axvline(x, color='cyan', linewidth=0.5)
        plt.axis('off')
        plt.title("Original Image with Grid")
        plt.savefig(os.path.join(output_directory, "original_image_with_grid.png"))
        plt.close()
        
        # Visualize Normalized Entropy
        fig, ax = plt.subplots(figsize=(12, 12))
        ax.imshow(image_array, cmap='gray', extent=[0, width, height, 0])
        for i, row in enumerate(normalized_entropy_grid):
            for j, entropy_value in enumerate(row):
                x = j * cell_size_px_x + cell_size_px_x // 2
                y = i * cell_size_px_y + cell_size_px_y // 2
                ax.text(x, y, f"{entropy_value:.2f}", color='red', fontsize=8, ha='center', va='center')
        for y in range(0, height, cell_size_px_y):
            ax.axhline(y, color='cyan', linewidth=0.5)
        for x in range(0, width, cell_size_px_x):
            ax.axvline(x, color='cyan', linewidth=0.5)
        plt.axis('off')
        plt.title("Normalized Entropy Visualization")
        plt.savefig(os.path.join(output_directory, "normalized_entropy.png"))
        plt.close()
        
        # Visualize Non-Normalized Entropy
        # Save Non-Normalized Entropy Visualization
        fig, ax = plt.subplots(figsize=(12, 12))
        ax.imshow(image_array, cmap='gray', extent=[0, width, height, 0])
        for i, row in enumerate(non_normalized_entropy_grid):
            for j, entropy_value in enumerate(row):
                x = j * cell_size_px_x + cell_size_px_x // 2
                y = i * cell_size_px_y + cell_size_px_y // 2
                ax.text(x, y, f"{entropy_value:.2f}", color='blue', fontsize=8, ha='center', va='center')
        for y in range(0, height, cell_size_px_y):
            ax.axhline(y, color='cyan', linewidth=0.5)
        for x in range(0, width, cell_size_px_x):
            ax.axvline(x, color='cyan', linewidth=0.5)
        plt.axis('off')
        plt.title("Non-Normalized Entropy Visualization")
        plt.savefig(os.path.join(output_directory, "non_normalized_entropy.png"))
        plt.close()
    
    return normalized_entropy_grid_pre, non_normalized_entropy_grid_pre

image_path = r"C:\Users\abuoliem\Box\3.Third year2025\1.Fall25\3.Research\5.Phase1_Damage paper\D.google earth pro\Pre image_1-8-20\Location1\50x50\retinex_corrected_sigma_30.jpg"
output_directory = r"C:\Users\abuoliem\Box\3.Third year2025\1.Fall25\3.Research\5.Phase1_Damage paper\D.google earth pro\Pre image_1-8-20\Location1\50x50"

cell_size_x = 50  # Cell size in meters (width)
cell_size_y = 50  # Cell size in meters (height)
resolution_x = 0.07  # Spatial resolution in meters/pixel (X-axis)
resolution_y = 0.07  # Spatial resolution in meters/pixel (Y-axis)

# Calculate entropy grids
normalized_entropy_grid_pre, non_normalized_entropy_grid_pre = analyze_entropy(
    image_path, output_directory, cell_size_x, cell_size_y, resolution_x, resolution_y, visualize=True
)    
```

```python
#---------------------------------------------- deeplabv3 model validation to confirm correct classification--------------------------------------------
#----------------------------------- 5 highest entropy values dynamically and validating specific entropy values simultaneously --------------------------------------------
#-----------------------------------                   without defining entropy threshold values --------------------------------------------

import time  # Import time module for execution time tracking
import numpy as np
import torch
import matplotlib.pyplot as plt
from PIL import Image
import torchvision.transforms as T
from torchvision.models.segmentation import deeplabv3_resnet101, DeepLabV3_ResNet101_Weights

# Start total execution time tracking
start_time = time.time()

###########################################           OBJECT DETECTION & VALIDATION            ############################################################

# ✅ Function to extract a patch from an image
def extract_patch(image, indices, grid_shape, cell_size_pixels):
    """
    Extracts a patch from an image based on entropy grid coordinates.
    """
    width, height = image.size  # Get image dimensions

    # Convert entropy grid coordinates to pixel coordinates
    cell_x = int((indices[1] / grid_shape[1]) * width)  
    cell_y = int((indices[0] / grid_shape[0]) * height)  

    # Define patch region (ensuring it's within image bounds)
    left = max(0, cell_x - cell_size_pixels // 2)
    top = max(0, cell_y - cell_size_pixels // 2)
    right = min(width, cell_x + cell_size_pixels // 2)
    bottom = min(height, cell_y + cell_size_pixels // 2)

    return image.crop((left, top, right, bottom))

###########################################           RUN PATCH EXTRACTION & VALIDATION           ############################################################

# ✅ Assume `difference_grid_1` is already computed and available in memory

# ✅ Find the top 5 highest entropy values in difference_grid_1
num_top_patches = 5
flat_indices = np.argpartition(difference_grid_1.flatten(), -num_top_patches)[-num_top_patches:]
top_indices = np.array(np.unravel_index(flat_indices, difference_grid_1.shape)).T

# ✅ Define additional entropy threshold for validation
entropy_threshold = 0.2  # Adjustable threshold for validation
validation_indices = np.argwhere(difference_grid_1 >= entropy_threshold)

# ✅ Combine top high-entropy indices and threshold-based indices
selected_indices = np.vstack((top_indices, validation_indices))

if selected_indices.size == 0:
    print("⚠️ No patches found! Adjust the entropy range.")
else:
    print("🔍 Selected patches:", selected_indices)  

# ✅ Load post-disaster image
post_disaster_image_path = r"C:\Users\abuoliem\Box\2.Second year2024\2.Spring2025\1.Research\2.satellite image analysis of pre- and post-disaster images\D.google earth pro\Post image_7-9-20\Lake Charles_Post_7-9-20.jpg"
post_disaster_image = Image.open(post_disaster_image_path).convert("RGB")

# ✅ Define cell size in pixels (714 pixels per cell)
cell_size_pixels = 714

# ✅ Extract patches
patches = []
patch_locations = []
for idx, indices in enumerate(selected_indices):
    patch = extract_patch(post_disaster_image, indices, difference_grid_1.shape, cell_size_pixels)
    patches.append(patch)
    patch_locations.append(indices)
    print(f"✅ Extracted patch {idx} at {indices} (Entropy Value: {difference_grid_1[indices[0], indices[1]]})")

###########################################           OBJECT SEGMENTATION MODEL (DeepLabV3)           ############################################################

# ✅ Load pre-trained DeepLabV3 model for segmentation
start_model_load_time = time.time()
segmentation_model = deeplabv3_resnet101(weights=DeepLabV3_ResNet101_Weights.DEFAULT)
segmentation_model.eval()
end_model_load_time = time.time()
print(f"⏱️ Model Loading Time: {end_model_load_time - start_model_load_time:.4f} seconds")

def segment_image(image):
    transform = T.Compose([
        T.Resize((512, 512)),  
        T.ToTensor(),
        T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    input_tensor = transform(image).unsqueeze(0)
    with torch.no_grad():
        output = segmentation_model(input_tensor)['out'][0]
    
    return output.argmax(0).byte().cpu().numpy()

###########################################           CLASSIFICATION           ############################################################

def classify_patch(segmentation_mask, entropy_value):
    unique_labels = np.unique(segmentation_mask)
    contains_building = 1 in unique_labels  
    
    print(f"🔍 Segmentation Labels: {unique_labels}")
    if contains_building and entropy_value > entropy_threshold:
        return "Building"
    else:
        return "Non-Building"

classified_patches = []
segmentation_masks = []

for idx, patch in enumerate(patches):
    entropy_value = difference_grid_1[patch_locations[idx][0], patch_locations[idx][1]]
    segmentation_mask = segment_image(patch)  
    segmentation_masks.append(segmentation_mask)
    classification = classify_patch(segmentation_mask, entropy_value)
    classified_patches.append((patch_locations[idx], classification))
    print(f"Patch {idx} at {patch_locations[idx]} Classification: {classification}")

###########################################           VISUALIZATION           ############################################################

fig, ax = plt.subplots(figsize=(10, 10))
im = ax.imshow(difference_grid_1, cmap='coolwarm', interpolation='none')
plt.colorbar(im, ax=ax, label='Entropy Difference')

ax.set_title('Entropy Difference Grid with Classification')
for loc, label in classified_patches:
    ax.text(loc[1], loc[0], label, ha='center', va='center', color='black', fontsize=8, bbox=dict(facecolor='white', alpha=0.6))

ax.axis('off')
plt.show()

print("🔍 Classified Patches:", classified_patches)

# ✅ End total execution time tracking
end_time = time.time()
print(f"⏱️ Total Execution Time: {end_time - start_time:.4f} seconds")
```
